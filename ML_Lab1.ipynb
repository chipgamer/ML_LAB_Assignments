```python
num=[2,7,4,1,3,6]
count=0
for i in range(0,6):
    for j in range(i+1,6):
        if num[i] + num[j]==10 :
            count=count+1
print(count)



def rng(lst, n):
    if n < 3:
        return False
   
    min = lst[0]
    max = lst[0]

   
    for x in range(1, n):
        if (lst[x] < min):
            min = lst[x]

   
    for y in range(1, n):
        if (lst[y] > max):
            max = lst[y]

   
    m = max - min
    return m


lst = []
n = int(input("Enter number of elements: "))
for p in range(0,n):
    ele = int(input())
    lst.append(ele)


result = rng(lst, n)
if result is False:
    print("Not enough elements to calculate range.")
else:
    print("Range:", result)
    
    

    
def matrix_multi(A, B):
    n=len(A)
    r=[[0]*n for _ in range(n)]
    for i in range(n):
        for j in range(n):
            for k in range(n):
                r[i][j]+=A[i][k]*B[k][j]
    return r
   

def power(A,m):
    r=A
    for _ in range(m - 1):
        r=matrix_multi(r, A)
    return r



A=[[4, 7, 2], [5, 3, 2], [9, 4, 5]]
m=int(input("enter value of m"))
r=power(A, m)

print("A^m=")
for i in r:
    print(i)
    
    
    
    
def highest_occurrence(s):
    count = {}
    for char in s:
        if char.isalpha():  # Consider only alphabets
            if char in count:
                count[char] += 1
            else:
                count[char] = 1

    max_char = max(count, key=count.get)
    max_count = count[max_char]

    return max_char, max_count

input_string = input("Enter a string: ")
char, count = highest_occurrence(input_string)
print(f"The highest occurring character is '{char}' with an occurrence count of {count}.")
```
